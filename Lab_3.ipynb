```python
#calculating weather through waterfall model
def calculate_coefficients(x, y):
    """Calculate the coefficients a, b, c for the quadratic model."""
    n = len(x)
    sum_x = sum(x)
    sum_y = sum(y)
    sum_x2 = sum(xi**2 for xi in x)
    sum_x3 = sum(xi**3 for xi in x)
    sum_x4 = sum(xi**4 for xi in x)
    sum_xy = sum(xi * yi for xi, yi in zip(x, y))
    sum_x2y = sum(xi**2 * yi for xi, yi in zip(x, y))
    det = (sum_x2 * sum_x4 * n + sum_x3 * sum_x2 * sum_x - sum_x3 * sum_x3 * n - sum_x2 * sum_x2 * sum_x2 - sum_x * sum_x4 * sum_x)
    det_a = (sum_x2 * sum_x4 * sum_y + sum_x3 * sum_x2 * sum_xy - sum_x3 * sum_x3 * sum_y - sum_x2 * sum_x2 * sum_xy - sum_x * sum_x4 * sum_y)
    det_b = (sum_x2 * sum_x4 * sum_xy + sum_x3 * sum_x2 * sum_y - sum_x3 * sum_x3 * sum_xy - sum_x2 * sum_x2 * sum_y - sum_x * sum_x4 * sum_y)
    det_c = (sum_x2 * sum_x4 * sum_xy + sum_x3 * sum_x2 * sum_y - sum_x3 * sum_x3 * sum_xy - sum_x2 * sum_x2 * sum_y - sum_x * sum_x4 * sum_y)
    a = det_a / det
    b = det_b / det
    c = det_c / det
    return a, b, c
def predict(x, a, b, c):
    """Predict temperature using the quadratic model."""
    return a * x**2 + b * x + c
def main():
    """Main function to implement the weather prediction system."""
    print("Welcome to the Weather Prediction System using Quadratic Regression!")
    n = int(input("Enter the number of data points (e.g., number of days): "))
    time = []
    temperature = []
    print(f"Please enter {n} pairs of (day, temperature) data:")
    for i in range(n):
        day = float(input(f"Enter day {i+1}: "))
        temp = float(input(f"Enter temperature for day {i+1} (°C): "))
        time.append(day)
        temperature.append(temp)
    print("\nCalculating quadratic model coefficients...")
    a, b, c = calculate_coefficients(time, temperature)
    print(f"\nQuadratic model coefficients:")
    print(f"a (quadratic term): {a}")
    print(f"b (linear term): {b}")
    print(f"c (constant term): {c}")
    predictions = [predict(day, a, b, c) for day in time]
    print("\nPredictions for the given days:")
    for day, pred in zip(time, predictions):
        print(f"Day {day}: Predicted Temperature = {pred:.2f} °C")
    print("\nYou can also predict the temperature for a future day.")
    future_day = float(input("Enter a future day to predict its temperature: "))
    future_pred = predict(future_day, a, b, c)
    print(f"Predicted Temperature for Day {future_day}: {future_pred:.2f} °C")
if __name__ == "__main__":
    main()
```
